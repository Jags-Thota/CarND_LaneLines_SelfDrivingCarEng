## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
rho = 1 # distance resolution in pixels of the Hough grid
theta = np.pi/180 # angular resolution in radians of the Hough grid
threshold = 15     # minimum number of votes (intersections in Hough grid cell)
min_line_len = 10 #minimum number of pixels making up a line
max_line_gap = 20    # maximum gap in pixels between connectable line segments

kernel_size=3
low_threshold=50
high_threshold=150
minSlopeThreshold=0.4

lower_yellow = np.array([22, 93, 0], dtype="uint8") # Lower  HSV value foe Yellow
upper_yellow = np.array([45, 255, 255], dtype="uint8") # Higher HSV value for yellow
lower_white = np.array([0,0,0], dtype=np.uint8)
upper_white = np.array([0,0,255], dtype=np.uint8)
````

````python
import math

def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) COLOR_RGB2GRAY
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

            
def draw_lines(img, lines, color=[255, 0, 0], thickness=10):

    rightLines, leftLines = [],[]
    drawRightLines=True
    drawLeftLines=True
    
    imageCenter_XCoord = img.shape[1]/2 # Image center 

    for line in lines:
        x1, y1, x2, y2 = line[0] 
        slope=(y2-y1)/(x2-x1) # calculate slope
        if np.isnan(float(slope)): # check division by 0 condition
            continue   
        if abs(slope)>minSlopeThreshold: # Filter out lines having slope less than the defined threshold
            if slope > 0 and x1>imageCenter_XCoord and x2>imageCenter_XCoord:  
                rightLines.append(line)
            elif slope < 0 and x1<imageCenter_XCoord and x2<imageCenter_XCoord:
                leftLines.append(line)
                
    # # Fit lines (y=Ax+B) usng the data points where x is slope
    rightXpoints, rightYpoints,leftXpoints,leftYpoints  =[],[],[],[] 
    
    #Right Lines
    for line in rightLines:
        x1, y1, x2, y2 = line[0]
        rightXpoints.extend([x1,x2])
        rightYpoints.extend([y1,y2])
    
    rightSlope, rightIntercept,leftSlope, leftIntercept=0,0,0,0
    
    if len(rightXpoints) > 0:
        rightSlope, rightIntercept = np.polyfit(rightXpoints, rightYpoints, 1)  # y = m*x + b
    else:
        drawRightLines = False
    
    # Left lane lines
    
    for line in leftLines:
        x1, y1, x2, y2 = line[0]
        leftXpoints.extend([x1,x2])
        leftYpoints.extend([y1,y2])

        
    if len(leftLines) > 0:
        leftSlope, leftIntercept = np.polyfit(leftXpoints, leftYpoints, 1)  # y = m*x + b
    else:
        drawLeftLines = False
                    
    # Find the end points 
    y1 = img.shape[0]
    y2 = img.shape[0] * 0.6
    
    if drawRightLines:
        rightX1 = (y1 - rightIntercept) / rightSlope
        rightX2 = (y2 - rightIntercept) / rightSlope
    if drawLeftLines:
        leftX1 = (y1 - leftIntercept) / leftSlope
        leftX2 = (y2 - leftIntercept) / leftSlope


    # Draw the right and left lines on image
    if drawRightLines:
        cv2.line(img, (int(rightX1), int(y1)), (int(rightX2), int(y2)), color, thickness)
    if drawLeftLines:
        cv2.line(img, (int(leftX1),  int(y1)), (int(leftX2), int(y2)), color, thickness)
    
    

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """  
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    #print(lines)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):

    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
def process_image(image):
    imshape = image.shape
    
    vertices = np.array([[(imshape[1]*0.20,imshape[0]),\
                      (image.shape[1]*0.4, image.shape[0]*0.6),\
                      (image.shape[1]*0.6, image.shape[0]*0.6),\
                      (imshape[1]*0.90,imshape[0])]]\
                    ,dtype=np.int32)


    grayImage = grayscale(image)
    hsvImage = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    # Convert the image to HSV then use color thresholding to identify Yellow lines
    mask_yellow = cv2.inRange(hsvImage, lower_yellow, upper_yellow)
    #maskW = cv2.inRange(hsvImage, lower_white, upper_white)
    mask_white = cv2.inRange(grayImage, 200, 255)
    mask_YW = cv2.bitwise_or(mask_white, mask_yellow)
    imageMask = cv2.bitwise_and(grayImage, mask_YW)
    #mpimg.imsave('test_images_output/Masked_'+ imageFile, imageMask) 
    blur_grayImage=gaussian_blur(imageMask, kernel_size)
    img_edges = canny(blur_grayImage, low_threshold, high_threshold)
    #mpimg.imsave('test_images_output/CannyEdges_'+ imageFile, img_edges)     
    masked_region= region_of_interest(img_edges, vertices)
    line_Image=hough_lines(masked_region, rho, theta, threshold, min_line_len, max_line_gap)
    #mpimg.imsave('test_images_output/Lines_'+ imageFile, line_Image) 

    #color_edges = np.dstack((img_edges, img_edges, img_edges))
    combo = cv2.addWeighted(image, 0.8, line_Image, 1, 0)  
    return combo
````
